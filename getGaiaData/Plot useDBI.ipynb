```python
#import packages 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#read in phot file
photData = pd.read_csv('NGC6819_5.phot',delim_whitespace=True)
df = pd.DataFrame(photData)
display(df)
```

```python
#plot pann stars (all stars in the file)
f, ax = plt.subplots(figsize=(8,10))
ax.plot(df['g_ps']-df['i_ps'],df['g_ps'],'.',color='deeppink')
ax.set_xlim(-0.5,3)
ax.set_ylim(23,10)
ax.set_xlabel('g_ps - i_ps',fontsize=18)
ax.set_ylabel('g_ps',fontsize=18)
# only plot useDBI=1
keep_var = 0
length= len(df.index)
print(length)
for i in range(length):
    if df.loc[keep_var,'useDBI']==1:
            ax.plot(df.loc[keep_var,'g_ps']-df.loc[keep_var,'i_ps'],df.loc[keep_var,'g_ps'],'.',color='blue')
       
    keep_var = keep_var + 1
```

```python

f, ax = plt.subplots(figsize=(8,10))
ax.plot(df['G_BP']-df['G_RP'],df['G'],'.',color='deeppink')
ax.set_xlim(-0.5,3)
ax.set_ylim(23,10)
ax.set_xlabel('G_BP - G_RP',fontsize=18)
ax.set_ylabel('G',fontsize=18)

keep_var = 0
length= len(df.index)
print(length)
for i in range(length):
    if df.loc[keep_var,'useDBI']==1:
            ax.plot(df.loc[keep_var,'G_BP']-df.loc[keep_var,'G_RP'],df.loc[keep_var,'G'],'.',color='blue')
       
    keep_var = keep_var + 1
```
